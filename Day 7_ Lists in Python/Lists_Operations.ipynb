## Operations on Lists

- Arithmetic
- Membership
- Loop

```python
L1 = [1, 2, 3, 4, 5]
L2 = [5, 6, 7, 8, 9]
```

```python
L1 + L2 #concatination
```

```python
L1 * 4
```

```python
# Membership

L1 = [1, 2, 3, 4, 5]
L2 = [5, 6, [7, 8, 9]]
```

```python
print(40 in L1)
```

```python
print(3 in L1)
```

```python
print([7, 8, 9] not in L2)
```

```python
# Loop
L1 = [1, 2, 3, 4, 5]

for i in L1:
    print(i)
```
