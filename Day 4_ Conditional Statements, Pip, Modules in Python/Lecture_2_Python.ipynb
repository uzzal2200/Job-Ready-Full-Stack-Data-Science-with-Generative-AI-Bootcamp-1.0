```python
# Login Function demo - dswithbapy.com 

""" 
Actual Credentials:

Email: entbappy73@gmail.com 
Password: Bappy@12345
"""

email = input("Enter your email: ")
password = input("Enter your password: ")


actual_email = "entbappy73@gmail.com"
actual_password = "Bappy@12345"

if email == actual_email and password == actual_password:
    print("Login Successfull!")

elif email == actual_email and password != actual_password:
    print("Incorrect Password!")


else:
    print("Invalid Credentials!")
```

# Nasted Conditional Statements

```python
# Login Function demo - dswithbapy.com 

""" 
Actual Credentials:

Email: entbappy73@gmail.com 
Password: Bappy@12345
"""

email = input("Enter your email: ")
password = input("Enter your password: ")


actual_email = "entbappy73@gmail.com"
actual_password = "Bappy@12345"

if email == actual_email and password == actual_password:
    print("Login Successfull!")

elif email == actual_email and password != actual_password:
    print("Please try again!")
    password = input("Enter your password again: ")

    if password == actual_password:
        print("Success!")
    
    else:
        print("Wrong!")


else:
    print("Invalid Credentials!")
```
