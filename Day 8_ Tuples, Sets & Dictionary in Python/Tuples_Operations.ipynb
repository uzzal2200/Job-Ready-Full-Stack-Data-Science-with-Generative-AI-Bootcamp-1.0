## Operations on Tuples

- Arithmetic
- Membership
- Loop

```python
# arithmetic operations +, *

t1 = (1,2,3,4)
t2 = (5,6,7,8)

t1 + t2 # concatination
```

```python
t1 * 4
```

```python
# membership

t1 = (1,2,3,4)
```

```python
1 in t1 
```

```python
1 not in t1
```

```python
# loop

t1 = (1,2,3,4)

for i in t1:
    print(i)
```
