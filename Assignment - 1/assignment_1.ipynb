#### Assignment 2: ATM Withdrawal System 

```python
# ATM Withdrawal System

# Get account balance from user
balance = float(input("Enter your account balance: $"))

# Get withdrawal amount from user
withdrawal = float(input("Enter the amount you want to withdraw: $"))

# Check if withdrawal is possible
if withdrawal > balance:
    print("Insufficient Balance")
else:
    print("Transaction Successful")
    remaining_balance = balance - withdrawal
    print(f"Remaining Balance: ${remaining_balance:.2f}")
```

#### Assignment 3: Student Grade Calculator 

```python
# Student Grade Calculator

# Get marks from user
marks = float(input("Enter your marks (0-100): "))

# Validate marks range
if marks < 0 or marks > 100:
    print("Invalid marks! Please enter a value between 0 and 100.")
else:
    # Determine grade based on marks
    if marks >= 90:
        grade = "A+"
    elif marks >= 80:
        grade = "A"
    elif marks >= 70:
        grade = "B"
    elif marks >= 60:
        grade = "C"
    else:
        grade = "Fail"

    # Display the result
    print(f"Your marks: {marks}")
    print(f"Your grade: {grade}")
```

#### Assignment 4: Bus Ticket Discount 

```python
# Bus Ticket Discount Calculator

# Define full ticket price
full_price = 100  # You can change this value

# Get age from user
age = int(input("Enter your age: "))

# Validate age
if age < 0:
    print("Invalid age! Please enter a valid age.")
else:
    # Determine ticket price based on age
    if age < 5:
        discount = "Free"
        ticket_price = 0
    elif age >= 5 and age <= 18:
        discount = "50% Discount (Child/Student)"
        ticket_price = full_price * 0.5
    elif age >= 60:
        discount = "30% Discount (Senior Citizen)"
        ticket_price = full_price * 0.7
    else:
        discount = "Full Price"
        ticket_price = full_price

    # Display the result
    print(f"\nAge: {age}")
    print(f"Category: {discount}")
    print(f"Ticket Price: ${ticket_price:.2f}")
```

#### Assignment 5: Restaurant Bill with Discount

```python
# Restaurant Bill with Discount

# Get bill amount from user
bill_amount = float(input("Enter total bill amount: BDT "))

# Check if discount applies
if bill_amount > 1000:
    discount = bill_amount * 0.10
    final_bill = bill_amount - discount
    print(f"\nOriginal Bill: BDT {bill_amount:.2f}")
    print(f"Discount (10%): BDT {discount:.2f}")
    print(f"Final Bill: BDT {final_bill:.2f}")
else:
    print(f"\nNo discount applied.")
    print(f"Final Bill: BDT {bill_amount:.2f}")
```

#### Assignment 8: Leap Year Checker  

```python

# Leap Year Checker

# Get year from user
year = int(input("Enter a year: "))

# Check if leap year
if year % 400 == 0:
    print(f"{year} is a Leap Year")
elif year % 4 == 0 and year % 100 != 0:
    print(f"{year} is a Leap Year")
else:
    print(f"{year} is Not a Leap Year")
```

#### Assignment 10: Electricity Bill Calculator

```python
# Electricity Bill Calculator

# Get units consumed from user
units = float(input("Enter number of units consumed: "))

# Calculate bill based on units
if units <= 100:
    bill = units * 5
elif units <= 200:
    bill = (100 * 5) + ((units - 100) * 7)
else:
    bill = (100 * 5) + (100 * 7) + ((units - 200) * 10)

# Display the bill
print(f"\nUnits Consumed: {units}")
print(f"Total Payable Amount: BDT {bill:.2f}")
```
