## Creating Stings

```python
s = 'hello'
s = "hello"
# multiline strings
s = '''hello'''
s = """hello"""
s = str('hello')
print(s)
```

```python
a = 34
str(a)
```

```python
text = "it's my bike"
print(text)
```
